```python
def genre_cov(o):
    d = []
    for i in ast.literal_eval(o):
        d.append(i['name'])
    return d
```

```python
movies.genres=movies.genres.apply(genre_cov)
```

```python
movies['genres']
```

```python
movies['keywords']=movies['keywords'].apply(genre_cov)
```

```python
## inn cast we are extractiong on first film characters
```

```python
def top_3(o):
    ct=0
    l=[]
    for i in ast.literal_eval(o):
        l.append(i['character'])
        ct=ct+1
        if ct==3:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(top_3)
```

```python
movies
```

```python
def director_name(o):
    l=[]
    for i in ast.literal_eval(o):
        if i['job']=="Director":
            l.append(i['name'])
            break
    return l
```

```python
"Director"
```

```python
movies['crew']=movies['crew'].apply(director_name)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split(','))
```

```python
def space_remover(o):
    d = []
    for i in o:
        l = []
        for j in i:
            if j != " ":
                l.append(j)
        d.append(''.join(l))
    return d
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
import nltk

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new['tags']=new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out().tolist()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```
